```python
fly = swarm.flies[8]
lrb = fly.experiments['lr_blob_expansion']
lrb.calc_spikepool()
```

```python
sp = lrb.sp
null_mask = ones_like(array(sp)).astype(np.bool)
pca_0 = sorters.PCATransform(sp,null_mask,defaults.pca_trans)
pca_0.transform()
km_0 = sorters.KMeansCluster(sp,null_mask,pca_0.trnsmtrx,defaults.km_cluster)
#km_0.params['init'] = km_test.est.cluster_centers_
km_0.select()
```

```python
sorters.plot_clusters(km_0,slice(0,4000,10))
display(gcf())
close()
```

```python
md_params = {'trans_dims':1}
mask = km_0.mask_from_labels('1')
md_01 = sorters.MedTrans(sp,mask,md_params)
md_01.transform()
th_01 = sorters.ThreshSelector(sp,mask,md_01.trnsmtrx,{})
thresh = percentile(md_01.trnsmtrx,98)
th_01.select(thresh)
```

```python
sorters.plot_clusters(th_01,slice(0,4000,10))
display(gcf())
close()
```
