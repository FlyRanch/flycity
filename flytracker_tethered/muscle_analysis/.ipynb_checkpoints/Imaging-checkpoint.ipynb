```python
import flylib
import tifffile
```

```python
abf_file = '/Volumes/FlyDataB/FlyDB/Fly0149/trial3_stripefix_14516007.abf'
tiff_file = '/Volumes/FlyDataB/FlyDB/Fly0149/trial3_stripefix.tif'
sigs = flylib.get_axon_signals(abf_file)
tif = tifffile.TiffFile(tiff_file)
images = tif.asarray()
```

```python
exposures = flylib.idx_by_thresh(sigs['CamSync'],thresh = 1)
frames = array([x[-1] for x in exposures])[0:-1].astype(int)
times = sigs['times']
```

```python
#they are fixating ok
xpos = sigs['Xpos']*2*pi/10
subplot(1,1,1,polar = True)
lns = hist(xpos,bins = 96)
display(gcf())
close()
```

```python
#time series of some ROI measurments I made in imageJ with the stripe position.
roi_data = genfromtxt("/Volumes/FlyDataB/FlyDB/Fly0149/trial3_roi.csv",delimiter=',',names = True)
fig = figure(figsize = (15,5))
ax0 = subplot(6,1,1)
plot(times[frames],xpos[frames])
axm1 = subplot(6,1,2,sharex = ax0)
plot(times[frames],roi_data['Meanm1'],color = 'k')
axm2 = subplot(6,1,3,sharex = ax0)
plot(times[frames],roi_data['Meanm2'],color = 'k')
axm3 = subplot(6,1,4,sharex = ax0)
plot(times[frames],roi_data['Meanm3'],color = 'k')
axm4 = subplot(6,1,5,sharex = ax0)
plot(times[frames],roi_data['Meanm4'],color = 'k')

ax0.set_ylabel('xpos (rad)')
axm1.set_ylabel('m1')
axm2.set_ylabel('m2')
axm3.set_ylabel('m3')
axm4.set_ylabel('m4')

ax0.set_xbound(95,120)
display(gcf())
#close()
```

```python
#find the correlation image between pixel fluorescence and a the left-right signal (might be better to do this with the right
#wing amplitude)
#first fiter the signal a bit.
import scipy.signal
lmr = sigs['Ph1']
times = sigs['times']
lowfilt = flylib.butter_bandpass_filter(lmr,0.01,100,times[1]-times[0],order = 1)
plot(times,lowfilt)
plot(times,lmr)
gca().set_xbound(0,1)
sig = lowfilt[frames]
#sig = gradient(lowfilt)[frames]
display(gcf())
close()
```
